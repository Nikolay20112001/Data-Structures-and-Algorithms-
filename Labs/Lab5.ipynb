### Кривая коха:

```python
import turtle

t = turtle.Turtle()
size = 500
n = int(input('Введите глубину фрактала: '))
t.speed(1000000)
t.hideturtle()

def koh_line(size, n):
    if n == 0:
        t.forward(size)
    else:
        koh_line(size/3, n-1)
        t.left(60)
        koh_line(size/3, n-1)
        t.right(120)
        koh_line(size/3, n-1)
        t.left(60)
        koh_line(size/3, n-1)

start_time = time.perf_counter() 
koh_line(size, n)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))
turtle.done()
```

### Ковер Серпинского:

```python
import queue, turtle


n = int(input('Введите глубину фрактала: '))
carpet = 'C'

def draw(fractal):
        for i in fractal:
            if i == 'f':
                t.up()
                t.forward(length)
                t.down()
            elif i == 'C':
                t.begin_fill()
                t.forward(length)
                t.left(angle)
                t.forward(length)
                t.left(angle)
                t.forward(length)
                t.left(angle)
                t.forward(length)
                t.left(angle)
                t.end_fill()
            elif i == '+':
                t.right(angle)
            elif i == '-':
                t.left(angle)

def Carpet(carpet, n):
    rules = {"C": "CfCfC-f+C-f+C--f--C--f--C+f-C+f-", "f": "fff"}
    if n == 0:
        return carpet
    else:
        end_string = "".join(   rules[i] if i in rules else i for i in carpet)
        carpet = end_string
        return Carpet(carpet, n-1)


t = turtle.Turtle()
t.hideturtle()
t.screen.bgcolor('pink')
t.color('white')
t.up()
t.goto(400, -200)
t.speed(100000000)
t.down()
t.pensize(1)
length = 6
angle = 90
t.left(90)

start_time = time.perf_counter()
fractal = Carpet(carpet, n)
draw(fractal)
print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))
turtle.done()
```

### Дерево Пифагора

```python
import queue, turtle


n = int(input('Введите глубину фрактала: '))
tree = '0'
def P_tree(tree, n):
    if n == 0:
        return tree
    else:
        tree = tree.replace('1', '11')
        tree = tree.replace('0', '1[0]0')
        return P_tree(tree, n-1)

start_time = time.perf_counter()
fractal = P_tree(tree, n)
stack = queue.LifoQueue()

size0 = 4
size1 = 30

t = turtle.Turtle()
t.speed(10000)
t.left(90)
t.pensize(3)
t.hideturtle()
t.up()
t.sety(-300)
t.down()
for i in fractal:
    if i == '1':
        t.forward(size1)
    elif i == '0':
        t.forward(size0)
    elif i == '[':
        x, y, angle = t.xcor(), t.ycor(), t.heading()
        stack.put((x, y, angle))
        t.left(45)
    elif i == ']':
        x, y, angle = stack.get()
        t.up()
        t.setx(x)
        t.sety(y)
        t.setheading(angle)
        t.down()
        t.right(45)

print("--- {:.3f} ms ---".format(((time.perf_counter() - start_time)*1000)))
turtle.done()
```
