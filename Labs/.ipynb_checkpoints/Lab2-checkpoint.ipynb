```python
import random
import time


#Бинарный поиск
def binary_search(lys, val):
    first = 0
    last = len(lys)-1
    index = -1
    while (first <= last) and (index == -1):
        mid = (first+last)//2
        if lys[mid] == val:
            index = mid
        else:
            if val<lys[mid]:
                last = mid -1
            else:
                first = mid +1
    return index

def binary_remove(mas, el):
    pos = binary_search(mas, el)
    x = mas
    del x[pos]
    return x

def get_random_list(n, min, max):
    s = [0] * n
    for i in range(n):
        s[i] = random.randint(min, max)
    return s

def binary_add(mas, el):
    x = mas
    x.append(el)
    x.sort()
    return x

s = sorted(get_random_list(10, 0, 10))
print('Исходный массив')
print(s)
print()
el = s[5]
start_time = time.time()
print('Ищем индекс 5 элемента (или равного ему) и проверяем ответ')
print(binary_search(s, el))
time_binary_search = round((time.time() - start_time)*1000)
print()
print('Удаляем 5 элемент')
print(binary_remove(s, el))
print()
print('Добавляем 100')
print(binary_add(s, 100))
```

```python
import random


class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.left = None
        self.right = None

    def __eq__(self, other):
        if isinstance(other, Node):
            return self.key == other.key

    def __lt__(self, other):
        if isinstance(other, Node):
            return self.key < other.key

    def __le__(self, other):
        if isinstance(other, Node):
            return self.key <= other.key

    def __gt__(self, other):
        if isinstance(other, Node):
            return self.key > other.key

    def __ge__(self, other):
        if isinstance(other, Node):
            return self.key >= other.key

    def hasLeft(self):
        return self.left != None

    def hasRight(self):
        return self.right != None

    def setleft(self, node):
        self.left = node

    def setright(self, node):
        self.right = node

    def setcurrent(self, key, value):
        self.key = key
        self.value = value


class Tree:

    def __init__(self, root_key, root_value):
        self.root = Node(root_key, root_value)

    def add(self, new_key, new_value):
        new_node = Node(new_key, new_value)
        current_pos = self.root
        isWhereToGo = True
        while isWhereToGo:
            if new_node < current_pos:
                if current_pos.hasLeft():
                    next_pos = current_pos.left
                    current_pos = next_pos
                else:
                    isWhereToGo = False
            else:
                if current_pos.hasRight():
                    next_pos = current_pos.right
                    current_pos = next_pos
                else:
                    isWhereToGo = False
        if current_pos == new_node:
            current_pos.setcurrent(new_key, new_value)
        elif new_node < current_pos:
            current_pos.setleft(new_node)
        else:
            current_pos.setright(new_node)

    def find_el(self, el_key):
        current_pos = self.root
        isWhereToGo = True
        check_node = Node(el_key, None)
        while isWhereToGo and current_pos.key != el_key:
            if check_node < current_pos:
                if current_pos.hasLeft():
                    next_pos = current_pos.left
                    current_pos = next_pos
                else:
                    isWhereToGo = False
            else:
                if current_pos.hasRight():
                    next_pos = current_pos.right
                    current_pos = next_pos
                else:
                    isWhereToGo = False
        if current_pos.key == el_key:
            return current_pos.value
        else:
            return -1

    def del_el(self, el_key):
        current_pos = self.root  # Находим нужную вершину
        steps = []
        isWhereToGo = True
        check_node = Node(el_key, None)
        while isWhereToGo and current_pos.key != el_key:
            if check_node < current_pos:
                if current_pos.hasLeft():
                    next_pos = current_pos.left
                    current_pos = next_pos
                    steps.append('left')
                else:
                    isWhereToGo = False
            else:
                if current_pos.hasRight():
                    next_pos = current_pos.right
                    current_pos = next_pos
                    steps.append('right')
                else:
                    isWhereToGo = False

        if not current_pos.hasRight():  # Если нет правого узла
            if current_pos.hasLeft():
                left_son = current_pos.left
                print(isinstance(left_son, Node))
                current_pos.key = left_son.key
                current_pos.value = left_son.value
                current_pos.left = left_son.left
                current_pos.right = left_son.right

            else:  # Находим узел-потомок текущего узла
                previous = self.root
                last_step = steps[-1]
                for i in range(0, len(steps) - 1):
                    if steps[i] == 'left':
                        next = previous.left
                        previous = next
                    elif steps[i] == 'right':
                        next = previous.right
                        previous = next

                current_pos.key = previous.key
                current_pos.value = previous.value
                if last_step == 'left':
                    previous.left = None
                elif last_step == 'right':
                    previous.right = None
                current_pos.left = previous.left
                current_pos.right = previous.right
        elif current_pos.hasRight():
            before_min = Node(current_pos.right.key, current_pos.right.value)
            before_min.left = current_pos.right.left
            before_min.right = current_pos.right.right
            stepsToBeforeMinn = 0
            if before_min.hasLeft():
                minn = before_min.left
                while minn.hasLeft():
                    stepsToBeforeMinn += 1
                    before_min.key = minn.key
                    before_min.value = minn.value
                    before_min.left = minn.left
                    before_min.right = minn.right
                    minn = minn.left

                current_pos.key = minn.key
                current_pos.value = minn.value

                #  Спускаемся чтобы удалить минимум, который был перенесен наверх
                cur_pos = current_pos.right
                for i in range(stepsToBeforeMinn):
                    next2 = cur_pos.left
                    cur_pos = next2

                cur_pos.left = None

                print(before_min.key)
                print(before_min.left)
                print(stepsToBeforeMinn)

            else:
                # Если у правого сына нет левого поддерева, то он сам является минимумом и мы "сдвигаем"
                # правое поддерево на место удаляемого элемента, при этом оставляю левое поддерево удаляемого элемента
                current_pos.key = current_pos.right.key
                current_pos.value = current_pos.right.value
                current_pos.right = current_pos.right.right




def random_tree(size, min, max):
    s = []
    for i in range(10):
        s.append(random.randint(min, max))
    print('mas:', s)
    tree = Tree(s[0], 0)
    for i in range(1, len(s)):
        tree.add(s[i], i)
    return tree

print('Находим индекс элемента 10 в случайном дереве на основе массива')

tree = random_tree(10, 1, 10)
start_time = time.time()
print(tree.find_el(10))
time_tree_search = round((time.time() - start_time)*1000)
print()

'''
Создаем наше дерево-тест c корневым узлом 33
                 (33, 0)
    (5, 1)
    
(1, 2)   (20, 3)

            (31, 4)
'''

tree = Tree(33, 0)
tree.add(5, 1)
tree.add(1, 2)
tree.add(20, 3)
tree.add(31, 4)

print('Проверяем как заполнилось дерево-тест:')

print(tree.root.key)
print(tree.root.left.key)
print(tree.root.left.left.key)
print(tree.root.left.right.key)
print(tree.root.left.right.right.key)

print()
print('Находим значение элемента с ключом 31:')
print(tree.find_el(31))
print()

print('Удаляем элемент под ключом 5 и проверяем наше дерево:')
print()

tree.del_el(5)
print(tree.root.key)
print(tree.root.left.key)
print(tree.root.left.left.key)
print(tree.root.left.right.key)
print(tree.root.left.right.right)
```

```python
#Метод Фибоначчи
def fibonacci_search(lst, target):
    size = len(lst)

    start = -1

    fm2 = 0
    fm1 = 1
    f = 1
    while (f < size):
        fm2 = fm1
        fm1 = f
        f = fm1 + fm2

    while (f > 1):
        index = min(start + fm2, size - 1)
        if lst[index] < target:
            f = fm1
            fm1 = fm2
            fm2 = f - fm1
            start = index
        elif lst[index] > target:
            f = fm2
            fm1 = fm1 - fm2
            fm2 = f - fm1
        else:
            return index
    if (fm1) and (lst[size - 1] == target) and index < len(lst) - 1:
        return size - 1
    return -1

def get_random_list(n, min, max):
    s = [0] * n
    for i in range(n):
        s[i] = random.randint(min, max)
    return s

def fibonacci_add(mas, el):
    x = mas
    x.append(el)
    x.sort()
    return x

def fibonacci_remove(mas, el):
    pos = binary_search(mas, el)
    x = mas
    del x[pos]
    return x

s = sorted(get_random_list(10, 0, 10))
print('Исходный массив')
print(s)
print()
el = s[5]
start_time = time.time()
print('Ищем индекс 5 элемента (или равного ему) и проверяем ответ')
print(fibonacci_search(s, el))
time_fibonacci_search = round((time.time() - start_time)*1000)
print()
print('Удаляем 5 элемент')
print(fibonacci_remove(s, el))
print()
print('Добавляем 100')
print(fibonacci_add(s, 100))
```

```python
#Интерполяционный
def InterpolationSearch(lys, val):
    low = 0
    high = (len(lys) - 1)
    while low <= high and val >= lys[low] and val <= lys[high]:
        index = low + int(((float(high - low) / ( lys[high] - lys[low])) * ( val - lys[low])))
        if lys[index] == val:
            return index
        if lys[index] < val:
            low = index + 1;
        else:
            high = index - 1;
    return -1


def get_random_list(n, min, max):
    s = [0] * n
    for i in range(n):
        s[i] = random.randint(min, max)
    return s

def InterpolationSearch_add(mas, el):
    x = mas
    x.append(el)
    x.sort()
    return x

def InterpolationSearch_remove(mas, el):
    pos = InterpolationSearch(mas, el)
    x = mas
    del x[pos]
    return x

s = sorted(get_random_list(10, 0, 10))
print('Исходный массив')
print(s)
print()
el = s[5]
start_time = time.time()
print('Ищем индекс 5 элемента (или равного ему) и проверяем ответ')
print(InterpolationSearch(s, el))
time_InterpolationSearch = round((time.time() - start_time)*1000)
print()
print('Удаляем 5 элемент')
print(InterpolationSearch_remove(s, el))
print()
print('Добавляем 100')
print(InterpolationSearch_add(s, 100))
```

```python
import random


#Простое рехэширование
class HashTable:
    def __init__(self, size=11):
        self.size = size
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self, key, data):
        hashvalue = self.hashfunction(key, len(self.slots))

        if self.slots[hashvalue] == None:
            self.slots[hashvalue] = key
            self.data[hashvalue] = data
        else:
            if self.slots[hashvalue] == key:
                self.data[hashvalue] = data  # replace
            else:
                nextslot = self.rehash(hashvalue, len(self.slots))
                while self.slots[nextslot] != None and \
                        self.slots[nextslot] != key:
                    nextslot = self.rehash(nextslot, len(self.slots))

                if self.slots[nextslot] == None:
                    self.slots[nextslot] = key
                    self.data[nextslot] = data
                else:
                    self.data[nextslot] = data  # replace

    def hashfunction(self, key, size):
        return key % size

    def rehash(self, oldhash, size):
        return (oldhash + 1) % size

    def get(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        data = None
        stop = False
        found = False
        position = startslot
        while self.slots[position] != None and \
                not found and not stop:
            if self.slots[position] == key:
                found = True
                data = self.data[position]
            else:
                position = self.rehash(position, len(self.slots))
                if position == startslot:
                    stop = True
        return data

    def remove(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        found = False
        position = startslot
        while self.slots[position] != None and \
                not found:
            if self.slots[position] == key:
                found = True
                self.slots[position] = None
                self.data[position] = None
            else:
                position = self.rehash(position, len(self.slots))


    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)
        

print('Формируем хеш-таблицу:')
print()
H = HashTable(4)
H[54]="cat"
H[26]="dog"
H[93]="lion"
H[17]="tiger"

print('keys:', H.slots)
print('values:', H.data)
print()

print('Удаляем элемент с ключом 54:')
print()

H.remove(54)
print('keys:', H.slots)
print('values:', H.data)
print()


print('Добавляем элемент с ключом 2 в массив:')
print()
H[2] = "new"
print('keys:', H.slots)
print('values:', H.data)
print()

print('Ищем элемент с ключом 17')
print()
print(H.get(17))
```

```python
#Рехэширование с помощью псевдослучайных чисел
import random


class HashTable2:
    def __init__(self, size=11):
        self.size = size
        self.slots = [None] * self.size
        self.data = [None] * self.size

    def put(self, key, data):
        hashvalue = self.hashfunction(key, len(self.slots))
        random.seed(hashvalue)
        pos = int(random.random() * self.size)

        if pos == self.size:
            pos -= 1

        if self.slots[pos] == None:
            self.slots[pos] = key
            self.data[pos] = data
        else:
            if self.slots[pos] == key:
                self.data[pos] = data  # replace
            else:
                # pseudorandom probing

                nextpos = int(random.random() * self.size)

                if nextpos == self.size:
                    pos -= 1

                while self.slots[nextpos] != None and \
                        self.slots[nextpos] != key:
                    nextpos = int(random.random() * self.size)

                    if nextpos == self.size:
                        nextpos -= 1

                if self.slots[nextpos] == None:
                    self.slots[nextpos] = key
                    self.data[nextpos] = data
                else:
                    self.data[nextpos] = data  # replace

    def hashfunction(self, key, size):
        return key % size

    def get(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        data = None
        stop = False
        found = False


        random.seed(startslot)
        start_pos = int(random.random() * self.size)
        position = start_pos

        # pseudorandom probing

        if position == self.size:
            position -= 1

        while self.slots[position] != None and \
                not found and not stop:
            if self.slots[position] == key:
                found = True
                data = self.data[position]
            else:
                position = int(random.random() * self.size)

                if position == self.size:
                    position -= 1

                if position == start_pos:
                    stop = True
        return data

    def remove(self, key):
        startslot = self.hashfunction(key, len(self.slots))

        random.seed(startslot)
        start_pos = int(random.random() * self.size)
        position = start_pos

        found = False

        while self.slots[position] != None and \
                not found:
            if self.slots[position] == key:
                found = True
                self.slots[position] = None
                self.data[position] = None
            else:
                position = int(random.random() * self.size)

                if position == self.size:
                    position -= 1


    def __getitem__(self, key):
        return self.get(key)

    def __setitem__(self, key, data):
        self.put(key, data)



print('Формируем хеш-таблицу:')
print()
H = HashTable2(4)
H[54]="cat"
H[26]="dog"
H[93]="lion"
H[17]="tiger"

print('keys:', H.slots)
print('values:', H.data)
print()

print('Удаляем элемент с ключом 54:')
print()

H.remove(54)
print('keys:', H.slots)
print('values:', H.data)
print()


print('Добавляем элемент с ключом 2 в массив:')
print()
H[2] = "new"
print('keys:', H.slots)
print('values:', H.data)
print()

print('Ищем элемент с ключом 17')
print()
print(H.get(17))
```

```python
#Метод цепочек
class chainHash:
    def __init__(self, size):
        self.keys = [None] * size
        self.datas = [None] * size
        self.size = size

    def put(self, key, data):
        hashvalue = self.hashfunction(key)
        if self.keys[hashvalue] == None:
            self.keys[hashvalue] = [key]
            self.datas[hashvalue] = [data]
        else:
            if key in self.keys[hashvalue]:

                for i in range(len(self.keys[hashvalue])):
                    if self.keys[hashvalue][i] == key:
                        self.datas[hashvalue][i] = data
            else:
                self.keys[hashvalue].append(key)
                self.datas[hashvalue].append(data)

    def get(self, key):
        hashvalue = self.hashfunction(key)
        if self.keys[hashvalue] == None:
            return None
        else:
            for i in range(len(self.keys[hashvalue])):
                if self.keys[hashvalue][i] == key:
                    return self.datas[hashvalue][i]
        return None

    def remove(self, key):
        hashvalue = self.hashfunction(key)
        if self.keys[hashvalue] != None:
            for i in range(len(self.keys[hashvalue])):
                if self.keys[hashvalue][i] == key:
                    if len(self.keys[hashvalue]) > 1:
                        self.datas[hashvalue] = self.datas[hashvalue][0:i] + self.datas[hashvalue][i + 1:]
                    else:
                        self.keys[hashvalue] = None
                        self.datas[hashvalue] = None


    def hashfunction(self, key):
        return key % self.size

    def __setitem__(self, key, data):
        self.put(key, data)

    def __getitem__(self, key):
        return self.get(key)


h = chainHash(2)
print('Создаем пустую хеш-таблицу:')
print()
print('keys:', h.keys)
print('values:',h.datas)
print()

print('Добавляем первые элементы:')
h[0] = 'a'
h[1] = 'c'
print()
print('keys:', h.keys)
print('values:', h.datas)
print()
print('Добавляем элемент с ключом 3 в заполненный массив:')
h[3] = 'd'
print()
print('keys:', h.keys)
print('values:', h.datas)
print()
print('Заменяем элемент с ключом 1')
h[1] = 'FF'
print()
print('keys:', h.keys)
print('values:', h.datas)
print()

print('удаляем элемент с ключом 0')
h.remove(0)
print()
print('keys:', h.keys)
print('values:', h.datas)
print()
```

### Задание №3
Расставить на стандартной 64-клеточной шахматной доске 8 ферзей так, чтобы ни один из них не находился под боем другого». Подразумевается, что ферзь бьёт все клетки, расположенные по вертикалям, горизонталям и обеим диагоналям
Написать программу,  которая находит хотя бы один способ решения задач.


```python
board = [[0 for i in range(8)] for j in range(8)]
count = 0

def setQueen(i, j):
    for x in range(8):
        board[x][j] += 1
        board[i][x] += 1
        if 0 <= i+j-x < 8:
            board[i+j-x][x] += 1
        if 0 <= i-j+x < 8:
            board[i-j+x][x] += 1
    board[i][j] = -1
    
def removeQueen(i, j):
    for x in range(8):
        board[x][j] -= 1
        board[i][x] -= 1
        if 0 <= i+j-x < 8:
            board[i+j-x][x] -= 1
        if 0 <= i-j+x < 8:
            board[i-j+x][x] -= 1
    board[i][j] = 0
    
def printPosition():
    global count
    count += 1
    abc = 'abcdefgh'
    ans = []
    for i in range(8):
        for j in range(8):
            if board[i][j] == -1:
                ans.append(abc[j]+str(i+1))
    print(', '.join(ans))
    
def solve(i):
    for j in range(8):
        if board[i][j] == 0:
            setQueen(i, j)
            if i == 7:
                printPosition()
            else:
                solve(i+1)
            removeQueen(i, j)
            
print("------------------Варианты расстановок------------------")
solve(0)
print("------------------Найденное количество расстановок------------------")
print(count)
```
